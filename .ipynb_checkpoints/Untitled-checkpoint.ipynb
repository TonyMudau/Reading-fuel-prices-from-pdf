```python
# importing required modules
from tabula import read_pdf
import pandas as pd
import matplotlib.pyplot as plt
```

```python

pd.set_option('display.max_rows', 500)
```

```python

def plot_fuel(file):
    df = read_pdf(r"C:\Users\Tony\Desktop\automations\{file}.pdf".format(file = file))

    fuelprices = df[0]
    price = fuelprices[5:]
    price.columns = price.iloc[0]
    gg = price.iloc[2:14,:].rename(columns={'c/litre': 'Month'})
    gg.reset_index(drop=True, inplace=True)
    gg = gg.set_index('Month')
    return plt.plot(gg['Unleaded'].astype(float))
```

```python
plot_fuel('2019fuel')
```

```python
plot_fuel('2020fuel')
```

```python
plot_fuel('2021fuel')
```
